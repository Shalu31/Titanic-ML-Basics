#categorical encoding ---
convert categories to numerical
#1.Nominal Variable: have no numerical ordering or order relationship between them
#2.Ordinal Variable:categories have some order

#Approaches to Categorical Encoding
1.Label Encoding (applied for norminal and ordinal)
2.oridnal (applied for ordinal)
3.One hot Encoding(applied for norminal)

```python
import pandas as pd
df = pd.DataFrame({'city': ['Mumbai','Delhi','Bangalore','Chennai','Delhi'],'age': [34,45,66,32,45],'height': [123.4,44.5,55.6,66.7,78.8]})
df
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['city'] = le.fit_transform(df['city'])
df
```

```python
import pandas as pd
df = pd.DataFrame({'city': ['Mumbai','Delhi','Bangalore','Chennai','Delhi'],'age': [34,45,66,32,45],'air_quality': ['good','bad','average','average','good']})
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['city'] = le.fit_transform(df['city'])
df
```

```python
df['air_quality'] = df['air_quality'].map({'bad':1,'average':2,'good':3})
df
```

```python
import pandas as pd
df = pd.DataFrame({'city': ['Mumbai','Delhi','Bangalore','Chennai','Delhi'],'age': [34,45,66,32,45],'height': [123.4,44.5,55.6,66.7,78.8]})
df
```

```python
#apply one hot encoding by using pandas
import numpy as np
df1 = pd.get_dummies(df,dtype = np.int8)
df1
```

```python
df2 = df1.drop(columns = ['city_Delhi'])
df2
```

```python
X = df2.drop(columns = ['height'])
y = df2['height']
```

```python
from sklearn.linear_model import LinearRegression
L = LinearRegression()
L.fit(X,y)
```

```python
Y_predict = L.predict(X)
Y_predict
```
