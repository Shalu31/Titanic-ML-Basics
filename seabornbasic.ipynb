```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
course = ['phy','chem','math']
students = [70,30,80]
sns.barplot(x=course,y=students)
plt.show()
```

```python
sns.scatterplot(x=course,y=students)
plt.show()
```

```python
sns.scatterplot(x=[2,3,4,5], y=[5,6,7,8])
plt.xlabel("x value")
plt.ylabel("y value")
plt.title("scatter plot")
plt.show()
```

```python
area = [200,300,400,500]
cost = [500,600,700,800]
sns.lineplot(x=area,y= cost)
plt.xlabel("area")
plt.ylabel("cost")
plt.title("line plot")
plt.show()
```

```python
days = [10,20,30]
cost = [50,30,10]
sns.lineplot(x=days,y= cost)
plt.xlabel("days")
plt.ylabel("cost")
plt.title("line plot")
plt.show()
```
